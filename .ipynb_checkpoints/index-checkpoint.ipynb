# H1 hypothesis testing
H1: Organic is more expensive than conventional

H0: There is no significant difference

We have filtered the data to a particular year 2017 so we minimize the effect of price inflation through out the 4 years worth of data we have. We have taken data from different regions we will apply a Z-Test because of the large data set we have (T-Test small sample size)

```python
value1= sample_conventional_2017.mean() - sample_organic_2017.mean()
print(value1)
zstat , pval1 = stests.ztest(sample_conventional_2017, x2=sample_organic_2017, value=0, alternative='smaller')
print(round(pval1))
if pval1<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
zstat    
```

# Hypothesis 2 (ANOVA Test)
There are regions were more organic avocados are sold therefore the average price for the avocados is lower on that region otherwise it remains the aound the same price throught the country.
- H0 : The average price of avocados throught the regions has no differences
- H1 : The average price of avocados is different depending on the region


```python
#Anova Test
formula = 'AveragePrice ~ C(region)'
lm = ols(formula, df_subregions).fit()
table = sm.stats.anova_lm(lm, typ=2)
print(table)
```

```python
# Tukey test after Anova
mc = MultiComparison(df_subregions.AveragePrice, df_subregions.region)
mc_results = mc.tukeyhsd()
print(mc_results)
```

Null hypothesis is rejected to all regions
